```python
import pandas as pd


df = pd.read_csv(r'C:\Users\hengj\OneDrive\Documents\GitHub\Hudson-Bay-Capital-Assessment\.ipynb_checkpoints\cleaned.csv', parse_dates=['created_date', 'closed_date'])#parse the date columns correctly
import matplotlib.pyplot as plt
```

```python
df.info()
```

```python
missing_counts =df.isnull().sum()# Count missing values in each column
print("Missing values in each column:")
print(missing_counts)
```

```python
top_complaint = 5
complaint=df['complaint_type'].value_counts().nlargest(top_complaint)
complaint.plot(kind='barh')
plt.xlabel("Number of Complaints per Type")
plt.ylabel("Type of Complaints")
plt.title("Top 5 Complaint Types")
plt.show()
```

```python
month_counts = df['month_created'].value_counts()
month_counts.plot(kind='barh')
plt.xlabel("Number of Complaints per month")
plt.ylabel("Month of Complaint")
plt.title("Complaints per Month")
plt.show()
```

```python
borough_counts = df['borough'].value_counts()
borough_counts.plot(kind='barh')
plt.xlabel("Borough")
plt.ylabel("Number of Complaints per Borough (In millions)")
plt.title("Complaints by Borough")
plt.tight_layout()
plt.show()
```

```python
agency = df['agency'].value_counts()
agency.plot(kind='barh')
plt.xlabel("Number of Complaints per Agency (In millions)")
plt.ylabel("Agency Name")
plt.title("Complaints by Agency")
plt.show() 
```

```python
df['complaint_type'].value_counts()
```

```python
city_complaint = 5
complaint=df['city'].value_counts().nlargest(city_complaint)
complaint.plot(kind='barh')
plt.xlabel("Number of Complaints per City (In Millions)")
plt.ylabel("City of Complaints")
plt.title("Top 5 Cities with Complaints")
plt.show()
```

```python
df.set_index('created_date', inplace=True)  # Set the index to the created_date column
daily_counts = df.resample('D').size()  # Resample to daily frequency and count occurrences
daily_counts.plot(kind='line')
plt.xlabel("Date")
plt.ylabel("Number of Complaints")
plt.title("Daily Complaints Over Time")
plt.show()
```

```python
borough_complaint = 'MANHATTAN'
manhattan_df = df[df['borough'] == 'MANHATTAN']
status_counts = manhattan_df['status'].value_counts()
status_counts.plot(kind='barh')
plt.xlabel("Number of Complaints")
plt.ylabel("Status of Complaints in " + borough_complaint)
plt.title("Status of Complaints in" + borough_complaint)
plt.tight_layout()
plt.show()
```
